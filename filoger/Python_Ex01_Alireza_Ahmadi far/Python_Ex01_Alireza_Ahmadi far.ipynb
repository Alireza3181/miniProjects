## Q1. The growth of CO<sub>2</sub> emissions

*How have CO<sub>2</sub> emissions grown over time?*

he dataset contains entries for countries grouped into four income groups: low, lower-middle, upper-middle and high. i) How have the total annual CO<sub>2</sub> emissions of each income group varied over the years 1950 to 2021? ii) How has the per capita CO<sub>2</sub> emissions (i.e., the emissions per person) of these groups varied over the same period. Comment on how the contribution of each income group is different when measured in terms of total emissions and per capita emissions.

(متن فارسی سوال در فایل پروژه موجود است)

```python
# ADD SOLUTION HERE. ADD AS MANY CELLS AS YOU NEED.

#A

income_groups = ['Low-income countries', 'Lower-middle-income countries', 'Upper-middle-income countries', 'High-income countries']
filtered_df = df[df['country'].isin(income_groups)]

filtered_df = filtered_df[(filtered_df['year'] >= 1950) & (filtered_df['year'] <= 2021)]

grouped_df = filtered_df.groupby(['country', 'year'])['co2'] .sum().unstack()

plt.figure(figsize=(12, 6))


for income_group in income_groups :
  plt.plot(grouped_df.columns, grouped_df.loc[income_group])

plt.title('CO2 emissions of each income group varied over the years 1950 to 2021')
plt.xlabel('Year')
plt.legend()
plt.grid(True)
plt.show()
```

```python
# B

grouped_df = filtered_df.groupby(['country', 'year'])['co2_per_capita'] .sum().unstack()

plt.figure(figsize=(12, 6))


for income_group in income_groups :
  plt.plot(grouped_df.columns, grouped_df.loc[income_group])

plt.title('CO2 per capita emissions of each income group varied over the years 1950 to 2021')
plt.xlabel('Year')
plt.legend()
plt.grid(True)
plt.show()
```
